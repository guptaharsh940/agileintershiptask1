Retrieve data from the sheet "(Input) User IDs"

```python
import pandas as pd
team_data = pd.read_excel(io="Python Assignment.xlsx",sheet_name="(Input) User IDs",skiprows=10, nrows=22, usecols='D:G')
team_data
```

Retrieve data from sheet "(Input) Rigorbuilder RAW"

```python
performance_data = pd.read_excel(io="Python Assignment.xlsx",sheet_name="(Input) Rigorbuilder RAW",skiprows=7, nrows=22, usecols='C:G')
performance_data
```

Evaluate and Dump data to csv for "Leaderboard TeamWise"

```python
result = {}
count = 1
for i in team_data.index:
    temp = performance_data.loc[performance_data["uid"]==team_data['User ID'][i]]['total_statements']
    temp2 = performance_data.loc[performance_data["uid"]==team_data['User ID'][i]]['total_reasons']
    if(not temp.empty):
        temp = temp.iloc[0]
        temp2 = temp2.iloc[0]
        if(team_data['Team Name'][i] in list(result.keys())):
            result[team_data['Team Name'][i]]["Average Statement"] +=temp
            result[team_data['Team Name'][i]]["Count"] +=1
            result[team_data['Team Name'][i]]["Average Reason"] +=temp2

        else:
            result[team_data['Team Name'][i]] = {"Average Statement":temp}
            result[team_data['Team Name'][i]]["Count"] =1
            result[team_data['Team Name'][i]]["Average Reason"] =temp2

print(result)
for i in result.keys():
    result[i]['Average Reason'] /= result[i]['Count']
    result[i]['Average Statement'] /= result[i]['Count']
result1 = pd.DataFrame(result)
result1 = result1.transpose()
result1['Count'] = result1['Average Statement']+result1['Average Reason']
result1 = result1.sort_values(by='Count', ascending=False)
result1 = result1.assign(Rank=lambda x: x['Count'].rank(ascending=False))
result1 = result1.reset_index()
last_column = result1.pop(result1.columns[-1])
result1.insert(0, last_column.name, last_column)
result1.columns.values[1] = 'Team Name'
result1.pop(result1.columns[3])
result1.to_csv("Leaderboard TeamWise.csv",index = False)
result1
```

Evaluate and Dump data to csv for "Leaderboard Individual"

```python
result2 = performance_data
result2['sum'] = result2['total_statements']+result2['total_reasons']
result2['names'] = result2['name'].apply(str.lower)
result2 = result2.sort_values(by=['sum','names'], ascending=[False,True])
result2=result2.reset_index()
result2 = result2.reset_index()
result2['level_0']+=1
last_column = result2.pop(result2.columns[-1])
result2.insert(0, last_column.name, last_column)
result2.pop("index")
result2.pop("S No")
result2.pop("sum")
result2.pop("names")
result2.columns.values[0] = 'Rank'
result2.columns.values[1] = 'Name'
result2.columns.values[2] = 'UID'
result2.columns.values[3] = 'No. of Statments'
result2.columns.values[4] = 'No. of Reasons'
result2.to_csv("Leaderboard Individual.csv",index = False)
result2
```
